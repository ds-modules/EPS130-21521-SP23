Seismic waveforms and spectra are used to estimate earthquake source parameters using the far-field theory we have derived. As shown in class at low frequencies the displacement amplitude spectrum is proportional to the scalar seismic moment, and the corner frequency can be used to infer the fault dimension (area) and the stress drop. On December 4, 1998 a ML=4.1 earthquake occurred in North Berkeley. This event was widely felt and caused only very minor damage locally. In this exercise you will attempt to model the tangential component Fourier amplitude spectrum of this event to determine the scalar seismic moment, and corner frequency. These measurements will be used to estimate the area of fault rupture and the stress drop.

Ascii files with the time history (brkseis.txt) and Fourier amplitude spectrum (brkspec.txt) are provided for computer analysis. The data in these files have been instrument corrected, and give ground motion in CGS units. For brkseis.txt column 1 is time (s) and column 2 is the seismogram amplitude cm. For brkspec.txt column 1 is frequency (Hz) and column 2 is the Fourier Amplitude spectra (cm*s). They are comma separated values (csv). You can 'poach' code from previous assignments on reading and plotting the data.

```python
#import needed packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Read and plot the timeseries and spectral data


```python
#read in the data and plot
timeseries=pd.read_csv('brkseis.txt')
spectrum=pd.read_csv('brkspec.txt')

time=np.array(timeseries['time'])        #Units: s
tsamp=np.array(timeseries['amplitude'])  #Units: cm

plt.plot(time,tsamp,'-')
plt.xlabel('Time(s)')
plt.ylabel('Displacement(cm)')
plt.title('Figure 1: Berkeley Earthquake Timeseries')
plt.show()

freq=np.array(spectrum['frequency'])        #Units: s
spamp=np.array(spectrum['amplitude'])  #Units: cm

plt.loglog(freq,spamp,'-')
plt.xlabel('Frequency(Hz)')
plt.ylabel('Amplitude(cm*s)')
plt.title('Figure 2: Berkeley Earthquake Amplitude Spectrum')
plt.show()
```
